```python
data= pd.read_csv("/work/FAC/FBM/DBC/cdessim2/default/amaurya/Outputs/STARCH_gene_outputs/Melanoma_starch_k3/labels_starch_k3.csv")
# new data frame with split value columns
new = data["Unnamed: 0"].str.split("x", n = 1, expand = True)
# making separate first name column from new data frame
data["x"]= new[0]
# making separate last name column from new data frame
data["y"]= new[1]

# convert coordinates to float type
# round the values to the nearest integer using the round() function
data['x'] = data['x'].astype(float)
data['x'] = round(data['x'])
data['y'] = data['y'].astype(float)
data['y'] = round(data['y'])

subclone_0 = data.loc[data['0'] == 0, ['x', 'y']]
subclone_1 = data.loc[data['0'] == 1, ['x', 'y']]
subclone_2 = data.loc[data['0'] == 2, ['x', 'y']]
subclone_3 = data.loc[data['0'] == 3, ['x', 'y']]


fig, ax = plt.subplots(figsize=(12, 12))
plt.scatter(subclone_0['x'], subclone_0['y'], c='yellow', s = 30)
plt.scatter(subclone_1['x'], subclone_1['y'], c='red', s = 30)
plt.scatter(subclone_2['x'], subclone_2['y'], c='magenta', s = 30)
plt.scatter(subclone_3['x'], subclone_3['y'], c='green', s = 30)

plt.xlabel('x')
plt.ylabel('y')
plt.legend('0123')
```
